```python
import numpy as np
from scipy import signal # conv2D
from PIL import Image
from matplotlib import pyplot as plt
```

```python
#Opening pic and show:
img = Image.open("image2.jpg")
plt.imshow(img) #opening image in plot
plt.axis('off') #off axis
plt.title('Original Image') #adding title
plt.show() #showing plot
#display(img) #displaying images, simple and fast variant 
#img.show()
```

```python
#GrayScaling!:
grayscale = img.convert("L")
#grayscale.show() #1) changing format of pic from rgb to L for getting grayscale pic and show..

#2) grayscaling pic in plot by adding args in function imshow
plt.imshow(grayscale, cmap='gray', vmin=0, vmax=255) #opening image in plot
plt.axis('off') #off axis
plt.title('Grayscale Image') #adding title
plt.show()
#img.shape ???
```

```python
#Conv:
Gx = np.array([[0,1,0],[1,-4,1],[0,1,0]]) #matrix that making pic more contrast (?)
conv = signal.convolve2d(grayscale, Gx) #applying conv for pic
#print(G) #print matrix

plt.imshow(conv, cmap='gray', vmin=0, vmax=255) #opening image in plot + vmin vmax
plt.axis('off') #off axis
plt.title('Myself contrasted Image') #adding title
plt.show()

fig, axs = plt.subplots(2, 1, figsize = (10, 5)) #showng several plots nearby
axs[0].imshow(grayscale, cmap='gray')# as on lesson without additional opening image in plot, cmap = colormap
axs[1].imshow(conv, cmap = 'gray')
for ax in axs.ravel():
    ax.axis('off')
axs[0].set_title("Input Image")
axs[1].set_title("Output Image")
plt.show()
```
