## Theory

Pandas is a data analysis and manipulation package for tabular data, such as data stored in spreadsheets or databases. It is used by several other packages. In this lesson only basic level of understanding is required. For introduction into its syntax, consider looking up [pandas cheat sheet](https://pandas.pydata.org/Pandas_Cheat_Sheet.pdf) online.

Pandas DataFrame is two-dimensional table with labeled axes; columns are headers, and rows are indexes. It can dynamically changed and can contain heterogeneous data. One row of data is sometimes refered to as an observation.

In this lesson, we will be working with Iris database. It can either downloaded, imported from CSV file or imported from some packages. It includes data about three iris species with 50 samples each. It contains the follow information about the flowers:

 - Its species name
 - Sepal length cm
 - Sepal widthin cm
 - Petal length in cm
 - Petal width in cm

```python
import numpy as np
import pandas as pd
from sklearn import datasets
import seaborn
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

# Opening CSV file
try:
    dataframe = pd.read_csv("iris.csv")
except IOError:
    pass

# Loading dataset from sklearn package
iris = datasets.load_iris()
# Creating data frame for pandas
dataframe = pd.DataFrame(iris['data'], columns=iris['feature_names'])

# Download from the internet
csv_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
# Requires setting column names manually
col_names = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Class']
dataframe =  pd.read_csv(csv_url, names = col_names)

print(dataframe.head(10))
print(dataframe.describe())
```
